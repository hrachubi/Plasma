# Plasma in B

Simulation of charged particle in B

* Units are normalized
* Code is not optimized
* Two calculation methods: Euler & LeapFrog

/*grab magnetic field at current position*/
B=EvalB(x);

/*get new velocity at n+1*/
v2[0] = v[0] + q/m*B*v[1]*dt;
v2[1] = v[1] - q/m*B*v[0]*dt;

/*update position*/
x2[0] = x[0] + v2[0]*dt;
x2[1] = x[1] + v2[1]*dt;

/*push down*/
v[0]=v2[0];
v[1]=v2[1];

```python
import matplotlib.pyplot as plt
import math
import numpy as np
#%matplotlib inline
#r=mv/qb
B=8

vx=[1]
vy=[0]
q=1
m=1
maxx=25000
totaltime=20
dt=0.1

t=[0]
r=[0]
rl=m*(math.sqrt(vx[0]*vx[0]+vy[0]*vy[0]))/(q*B)
rll=[rl]
r=[rl]
dr=[0]
x=[rl]
y=[rl]


dt=totaltime/float(maxx)
print('larmor= ',rl,'dt= ',round(dt,4),' steps= ',maxx)

#/*grab magnetic field at current position*/
#B=EvalB(x);

#/*get new velocity at n+1*/
for i in range(1,maxx):
    #vx.append((vx[i-1]+i)*i)
    #print(i,vx)
    t.append(t[i-1]+dt)
    vx.append(vx[i-1] + q/m*B*vy[i-1]*dt)
    vy.append(vy[i-1] - q/m*B*vx[i-1]*dt)
    x.append(x[i-1]+vx[i]*dt)
    y.append(y[i-1]+vy[i]*dt)
    r.append(m*(math.sqrt(vx[i]*vx[i]+vy[i]*vy[i]))/(q*B))
    dr.append(r[i]-rl)
    rll.append(rl)
    
vtot = [math.sqrt(a**2+b**2) for a, b in zip(vx,vy)]

theta=np.linspace(0,np.pi*2,100)
xr=1/B+rl*np.sin(theta)
yr=rl*np.cos(theta)

plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.plot(x,y,label="calculation")
plt.plot(xr,yr,'r--',label='Theory')
plt.xlabel('X position')
plt.ylabel('Y position')
plt.title('Position')
plt.legend()


plt.subplot(2,3,2)
plt.plot(t,r,label='calc')
plt.plot(t,rll,'r--',label='theory')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Larmor')
plt.title('Larmor')


plt.subplot(2,3,3)
plt.plot(t,dr)
plt.xlabel('Time')
plt.ylabel('dr')
plt.title('d_Larmor')

plt.subplot(2,3,4)
plt.plot(t,vtot)
plt.xlabel('Time')
plt.title('Total speed')

plt.subplot(2,3,5)
plt.plot(t,vx)
plt.xlabel('Time')
plt.title('Vx speed')

plt.subplot(2,3,6)
plt.plot(t,vy)
plt.xlabel('Time')
plt.title('Vy speed')

plt.show()
#print(t,'\n',vx,'\n',vy)
```

```python
#LeapFrog

q=1
m=1
maxx=25000
totaltime=20
dt=0.1
dt=totaltime/float(maxx)
print('larmor= ',rl,'dt= ',round(dt,4),' steps= ',maxx)


vxLF=[1]
vyLF=[0]
tLF=[0]
rLF=[0]
#rllLF=[rl]
rl=m*(math.sqrt(vx[0]*vx[0]+vy[0]*vy[0]))/(q*B)
rLF=[rl]
drLF=[0]
xLF=[rl]
yLF=[rl]

vxLF.append(vxLF[0] + q/m*B*vyLF[0]*dt/2)
vyLF.append(vyLF[0] - q/m*B*vxLF[0]*dt/2)



#/*grab magnetic field at current position*/
#B=EvalB(x);

#/*get new velocity at n+1*/
for i in range(1,maxx):
    #vx.append((vx[i-1]+i)*i)
    #print(i,vx)
    tLF.append(tLF[i-1]+dt)
    xLF.append(xLF[i-1]+vxLF[i]*dt/2)
    yLF.append(yLF[i-1]+vyLF[i]*dt/2)
    vxLF.append(vxLF[i-1] + q/m*B*vyLF[i-1]*dt)
    vyLF.append(vyLF[i-1] - q/m*B*vxLF[i-1]*dt)

    rLF.append(m*(math.sqrt(vxLF[i]*vxLF[i]+vyLF[i]*vyLF[i]))/(q*B))
    drLF.append(rLF[i]-rl)
    
vxLF.pop()
vyLF.pop()
vtotLF = [math.sqrt(a**2+b**2) for a, b in zip(vxLF,vyLF)]


theta=np.linspace(0,np.pi*2,100)
xr=1/B+rl*np.sin(theta)
yr=rl*np.cos(theta)

plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.plot(xLF,yLF,label="calculation")
plt.plot(xr,yr,'r--',label='Theory')
plt.xlabel('X position')
plt.ylabel('Y position')
plt.title('Position')
plt.legend()


plt.subplot(2,3,2)
plt.plot(t,rLF,label='calc')
plt.plot(t,rll,'r--',label='Leap Frog')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Larmor')
plt.title('Larmor')


plt.subplot(2,3,3)
plt.plot(t,drLF)
plt.xlabel('Time')
plt.ylabel('dr')
plt.title('d_Larmor')

plt.subplot(2,3,4)
plt.plot(tLF,vtotLF)
plt.xlabel('Time')
plt.title('Total speed')

plt.subplot(2,3,5)
plt.plot(tLF,vxLF)
plt.xlabel('Time')
plt.title('Vx speed')

plt.subplot(2,3,6)
plt.plot(tLF,vyLF)
plt.xlabel('Time')
plt.title('Vy speed')

plt.show()
#print(t,'\n',vx,'\n',vy)
```
